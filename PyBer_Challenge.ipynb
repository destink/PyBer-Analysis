### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head(20)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_city_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_city_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers_city = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_city
```

```python
#  3. Get the total amount of fares for each city type
total_fares_city = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_city
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_ride = total_fares_city / total_city_rides
avg_fare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_driver_city = total_fares_city / total_drivers_city
avg_fare_driver_city
```

```python
#  6. Create a PyBer summary DataFrame. TODO
pyber_summary = {'Total Rides': total_city_rides, 'Total Drivers': total_drivers_city, 'Total Fares': total_fares_city, 'Average Fare per Ride': avg_fare_ride, 'Average Fare per Driver': avg_fare_driver_city}
pyber_summary_df = pd.DataFrame(pyber_summary)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
# Format the colums
for column in ['Total Fares', 'Average Fare per Ride', 'Average Fare per Driver']:
    pyber_summary_df[column] = pyber_summary_df[column].map('${:,.2f}'.format)
```

```python
# Format the colums
for column in ['Total Rides','Total Drivers']:
    pyber_summary_df[column] = pyber_summary_df[column].map('{:,}'.format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fares_df = pyber_data_df.groupby(['type', 'date']).sum()['fare']
sum_fares_df
```

```python
# 3. Reset the index on the DataFrame you created in #2. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fares_pivot_df = sum_fares_df.reset_index()
sum_fares_pivot_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_fares_pivot = sum_fares_pivot_df.pivot_table(index='date', columns='type', values='fare')
sum_fares_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
sum_fares_new_df = sum_fares_pivot.loc['2019-01-01':'2019-04-29', :]
sum_fares_new_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
sum_fares_new_df.index = pd.to_datetime(sum_fares_new_df.index)
sum_fares_new_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_fares_new_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = sum_fares_new_df.resample('W').sum()
weekly_fares_df.head()
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
weekly_fares_df.plot(figsize = (12,6))
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
plt.ylabel("Fare ($USD)")
plt.title("Total Fares by City Type")                     
plt.xlabel("Date")
plt.tight_layout()
plt.savefig("Resources/sum_fares_city.png")
plt.show()
```
